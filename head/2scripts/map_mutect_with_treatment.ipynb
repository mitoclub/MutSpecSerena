```python
import os, glob
import pandas as pd
```

Load data from every file with **Mutect2** results and add it to `mutect_df`:

```python
os.chdir("../../body/1raw/MtMutectAnnovar/")

mutect_df = pd.DataFrame(
    columns=['CHROM', 'POS', 'ID', 'REF', 'ALT', 'QUAL', 'FILTER', 'INFO', 'FORMAT', 'SAMPLE_ID', 'filename'])

for file in glob.glob("*mutect.mt"):
    df = pd.read_csv(file, sep='\t', header=None,  comment='#')
    df.columns = ['CHROM', 'POS', 'ID', 'REF', 'ALT', 'QUAL', 'FILTER', 'INFO', 'FORMAT', 'SAMPLE_ID']
    sample = file.split(".")[0]
    df['filename'] = sample
    mutect_df = mutect_df.append(df)
```

```python
print(mutect_df.shape[0], 'lines')
mutect_df.head()
```

Load the table with mapped info:

```python
info_df = pd.read_csv("../treatment_mapped_to_kostya_files.csv", index_col=0)
print(info_df.shape[0], 'lines')
info_df.head()
```

Merge the two tables into one `grand_df`:

```python
grand_df = pd.merge(mutect_df, info_df, how='outer', on='filename')
print(grand_df.shape[0], 'lines')
grand_df.head()
```

Save the table to a `.csv` file:

```python
grand_df.to_csv("../../2derived/mutations_vs_treatment.csv")
```
